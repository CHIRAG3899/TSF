```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

eda_df = pd.read_csv('SampleSuperstore.csv')
eda_df.head()
```

```python
eda_df.info()
```

```python
eda_df.describe()
```

```python
eda_df.corr()
```

```python
sns.heatmap(eda_df.corr() , annot=True , cbar=True)
```

```python
eda_df.isnull().sum()
```

```python
eda_df.nunique()
```

```python
fig , axs = plt.subplots(ncols = 2 , nrows = 2 , figsize = (10,10))
sns.distplot(eda_df['Sales'], color = 'blue', ax = axs[0][0])
sns.distplot(eda_df['Profit'], color = 'red', ax = axs[0][1])
sns.distplot(eda_df['Quantity'], color = 'green', ax = axs[1][0])
sns.distplot(eda_df['Discount'], color = 'orange', ax = axs[1][1])

axs[0][0].set_title('Sales distribution' , fontsize = 20)
axs[0][1].set_title('Profit distribution' , fontsize = 20)
axs[1][0].set_title('Quantity distribution' , fontsize = 20)
axs[1][1].set_title('Discount distribution' , fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(eda_df['Sales'],color='red',label='Sales',shade=True,bw=25)
sns.kdeplot(eda_df['Profit'],color='green',label='Profit',shade=True,bw=25)
plt.xlim([-100,1000])
plt.legend()
```

```python

eda_df['City'].unique()
eda_df['City'].value_counts
fig ,ax = plt.subplots(figsize = (30,8))
sns.countplot(eda_df['City'].head(100),ax = ax)
```

```python
eda_df_city = eda_df.groupby('City')[['Sales','Discount','Profit']].mean()
eda_df_city.head(10)
```

```python
eda_df_city.tail(10)
```

```python
#Sorting the cities with respect to profit in descending
eda_df_city = eda_df_city.sort_values('Profit',ascending = False)
#Top ten 10 Profit giving cities
eda_df_city.head(10)
```

```python
#Plot Profit,Discount ,sale of Top 30 cities
eda_df_city.head(30).plot(kind = 'bar',color = 'red',figsize= (18,20))
eda_df_city['Profit'].head(30).plot(kind='bar',figsize=(15,5),color = 'Green')
plt.title('City wise analysis of  profit')
```

```python
#Low Profit
eda_df_city['Profit'].tail(30).plot(kind='bar',figsize=(15,5),color = 'yellow')
plt.title('City wise analysis of profit')
```

```python
eda_df['State'].value_counts
eda_df['State'].unique()
eda_df_state = eda_df.groupby('State')['Sales','Discount','Profit'].mean()
eda_df_state.head()
```

```python
eda_df_state.sort_values('Profit',ascending = False,inplace = True)
eda_df_state.head(10)
```

```python
eda_df_state[['Profit']].plot(kind = 'bar',figsize = (20,8))
plt.xlabel('Profit ')
plt.ylabel('Name of States')
plt.title('Profit vs States')
plt.legend()
```

```python
eda_df_state[['Sales']].plot.pie(subplots = True,autopct = "%1.2f%%",figsize =(50,20),labels = eda_df_state.index)
```

```python
#Let plot graph of top h 30 ighest sales states
eda_df_state1 = eda_df_state.sort_values('Sales',ascending = False)
eda_df_state1[['Sales']].head(30).plot(kind = 'bar',color = 'orange',figsize = (20,8))
plt.xlabel('States',fontsize = 15)
plt.ylabel('Sales')
plt.title('Plot of Top 30 Sales State',fontsize = 20)
```

```python
#Highest sale : Wyoming
eda_df_state1[['Sales']].tail(30).plot(kind = 'bar',color = 'blue',figsize = (20,8))
plt.title('Plot of least Sales State',fontsize = 20)
plt.xlabel('States',fontsize =15)
plt.ylabel('Sales',fontsize = 15)
```

```python
eda_df_state1['Discount'].plot(kind='bar',figsize=(20,5))
plt.title('State wise analysis of Discount', fontsize=20)
plt.xlabel('State',fontsize = 15)
plt.ylabel('Discount',fontsize = 15)
```
